```python
from pymongo import MongoClient

dbName = "companies"
mongodbURL = f"mongodb://localhost/{dbName}"

client = MongoClient(mongodbURL, connectTimeoutMS=2000, serverSelectionTimeoutMS=2000)
db = client.get_database()
```

```python
query={"offices.state_code":"MI"}
offices=list(db.companies.find(query,{"name":1,"offices":1}))
companies=pd.DataFrame(offices)
allstates=companies.explode("offices").apply(lambda x: x.offices, axis=1, result_type="expand")
miami=allstates[allstates["state_code"]=="MI"]
mi=miami.rename(columns={"latitude":"lat","longitude":"lng"})
```

```python
mi["point"]=mi.apply(geo.TransformGeo, axis=1)
mi.head()
```

```python
mi.to_json("output/miamicompanies.json", orient="records")
```

```python
def geocode(address):
    res = requests.get(f"https://geocode.xyz/{address}", params={"json":1})
    data = res.json()
    return {
        "type":"Point",
        "coordinates":[float(data["longt"]),float(data["latt"])]
    }
times_square = geocode("Times Square, New York")
times_square
```
